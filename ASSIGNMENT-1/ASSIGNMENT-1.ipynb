```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

# Q1




```python
wiki_page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
wiki_content=bs(wiki_page.content)
wiki_headers=[]
for i in wiki_content.find_all('span',class_="mw-headline"):
    wiki_headers.append(i.text)
wiki_df=pd.DataFrame({"Headers":wiki_headers})
wiki_df
```

# Q2

```python
presidents_page=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
presidents_content=bs(presidents_page.content)
presidents_names=[]
for i in presidents_content.find_all('div',class_="presidentListing"):
    presidents_names_raw=[]
    presidents_names_raw.append(i.text.split('\n')[1])
    for i in presidents_names_raw:
        presidents_names.append(i.split('(')[0])
presidents_TOO=[]
for i in presidents_content.find_all('div',class_="presidentListing"):
    presidents_TOO_raw=[]
    presidents_TOO_raw.append(i.text.split('Office:')[1])
    for i in presidents_TOO_raw:
        presidents_TOO.append(i.split('\n')[0])
presidents_df=pd.DataFrame({"Names":presidents_names, "Term of Office":presidents_TOO})
presidents_df
```

# A

```python
odi_m_team_page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
odi_m_team_content=bs(odi_m_team_page.content)

breaker=0
jump=0
odi_m_team=[]
for i in odi_m_team_content.find_all('span', class_="u-hide-phablet"):
    breaker+=1
    if breaker==11:
        breaker=0
        break
    odi_m_team.append(i.text)

    
odi_m_matches=[]
firstmatch=odi_m_team_content.find('td',class_="rankings-block__banner--matches")
odi_m_matches.append(firstmatch.text)
for i in odi_m_team_content.find_all('td',class_="table-body__cell u-center-text"):
    jump+=1
    if jump % 2==0:
        continue
    odi_m_matches.append(i.text)
    breaker+=1
    if breaker==9:
        breaker=0  
        jump=0
        
        break

        
odi_m_points=[]
firstpoints=odi_m_team_content.find('td',class_="rankings-block__banner--points")
odi_m_points.append(firstpoints.text)
for i in odi_m_team_content.find_all('td',class_="table-body__cell u-center-text"):
    jump+=1
    if jump % 2!=0:
        continue
    odi_m_points.append(i.text)
    breaker+=1
    if breaker==9:
        breaker=0
        break
        
odi_m_ratings=[]       
firstratings=odi_m_team_content.find('td',class_="rankings-block__banner--rating u-text-right")
odi_m_ratings.append(firstratings.text.split("\n")[1])
for i in odi_m_team_content.find_all('td',class_="table-body__cell u-text-right rating"):
    breaker+=1
    if breaker==10:
        break
    odi_m_ratings.append(i.text)


odi_m_team_df=pd.DataFrame({"Teams":odi_m_team, "Matches":odi_m_matches, "Points":odi_m_points, "Ratings":odi_m_ratings})
odi_m_team_df
```

```python
odi_batsman_page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi")
odi_batsman_content=bs(odi_batsman_page.content)

temp=0
odi_batsman_name=[]
firstname=odi_batsman_content.find('div',class_="rankings-block__banner--name")
odi_batsman_name.append(firstname.text)
for i in odi_batsman_content.find_all('td', class_="table-body__cell name"):
    odi_batsman_name.append(i.text.split('\n')[1])
    temp+=1
    if temp==9:
        temp=0
        break
       

        
odi_batsman_team=[]
firstname=odi_batsman_content.find('div',class_="rankings-block__banner--nationality") 
odi_batsman_team.append(firstname.text.split('\n')[2])
for i in odi_batsman_content.find_all('span', class_="table-body__logo-text"):
    odi_batsman_team.append(i.text)
    temp+=1
    if temp==9:
        temp=0
        break
        
odi_batsman_ratings=[]
firstname=odi_batsman_content.find('div',class_="rankings-block__banner--rating")
odi_batsman_ratings.append(firstname.text)
for i in odi_batsman_content.find_all('td', class_="table-body__cell u-text-right rating"):
    odi_batsman_ratings.append(i.text)
    temp+=1
    if temp==9:
        break

odi_batsman_df=pd.DataFrame({"Names":odi_batsman_name, "Teams":odi_batsman_team, "Ratings":odi_batsman_ratings})
odi_batsman_df
```

```python
news_page=requests.get('https://www.cnbc.com/world/?region=world')
news_content=bs(news_page.content)


news_headlines=[]
for i in news_content.find_all('a',class_="LatestNews-headline"):
    news_headlines.append(i.text)

news_links=[]
for i in news_content.find_all('a',class_="LatestNews-headline"):
    news_links.append(i.get('href'))
news_links

news_time=[]
for i in news_content.find_all('time',class_="LatestNews-timestamp"):
    news_time.append(i.text)
    


news_df=pd.DataFrame({'Headlines':news_headlines,"Links":news_links,'Time':news_time})
news_df
```

```python
articles_page=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
articles_content=bs(articles_page.content)


titles=[]
for i in articles_content.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    titles.append(i.text)

    
authors=[]
for i in articles_content.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    authors.append(i.text)

    
date=[]
for i in articles_content.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    date.append(i.text)

    
url=[]
for i in articles_content.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    url.append(i.get('href'))


articles_df=pd.DataFrame({'Titles':titles,'Authors':authors,'Date':date,'URL':url})
articles_df
```

```python
restaurant_page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
restaurant_content=bs(restaurant_page.content)

name=[]
for i in restaurant_content.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text)

    
cuisine=[]
for i in restaurant_content.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])

    
location=[]
for i in restaurant_content.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
    
ratings=[]
for i in restaurant_content.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
    

image_url=[]
for i in restaurant_content.find_all('img',class_="no-img"):
    image_url.append(i.get('data-src'))

    
restaurant_df=pd.DataFrame({'Restaurant Names':name,'Cuisines':cuisine,"Location":location,'Ratings':ratings,'Image URL':image_url})
restaurant_df
```
