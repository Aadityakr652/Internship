```python
# Connecting to the driver
dr=webdriver.Chrome(executable_path=driver_path, options=options)
#Opening the website on automated brave browser
dr.get('http://www.naukri.com/')

#Sending the required keys
designation=dr.find_element(By.XPATH,'//*[@id="root"]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input')
designation.send_keys('Data Analyst') #Sending "Data Analyst" on designation field

location=dr.find_element(By.XPATH,'//*[@id="root"]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Bangalore') #Sending "Bangalore" on location field

#clicking the search button
button=dr.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[6]')
button.click()

#Creating empty lists for the required data
job_title=[] 
job_location=[] 
company_name=[] 
experience_required=[]

#Adding delay to allow the website to load and update the content
time.sleep(4)

#Scraping the required data
for i in dr.find_elements(By.XPATH,'//a[@class="title ellipsis"]')[0:10]: #Using slice method to scrape only first 10 datas
    job_title.append(i.text) # Appending Job title

for i in dr.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')[0:10]: #Using slice method to scrape only first 10 datas
    job_location.append(i.text) # Appending Job location


for i in dr.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')[0:10]: #Using slice method to scrape only first 10 datas
    company_name.append(i.text)  # Appending Company Name  

    
for i in dr.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')[0:10]: #Using slice method to scrape only first 10 datas
    experience_required.append(i.text)  # Appending Experience required
    
#Creating the Data Frame 
Df_1=pd.DataFrame({'Job Title':job_title,'Job Location':job_location,'Company Name':company_name,'Experience required':experience_required})
Df_1
```

```python
# Connecting to the driver
dr=webdriver.Chrome(executable_path=driver_path, options=options)
#Opening the website on automated brave browser
dr.get('https://www.flipkart.com/')

#Closing the login tab
dr.find_element(By.XPATH,"/html/body/div[2]/div/div/button").click()

#Sending the required key
search=dr.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search.send_keys('Sunglass') # Sending "Sunglass" key to the search field

#Clicking the search button
button=dr.find_element(By.CLASS_NAME,"L0Z3Pu")
button.click() 
```

```python
#Creating empty lists for the required data
Brand=[] 
Product=[] 
Price=[] 
Discount=[]
urls=[]

#Scraping the URL's of pages
for i in dr.find_elements(By.XPATH,'//a[contains(@class,"ge-49M")]')[0:3]: #Scraping ulrs of the pages and slicing them  
        urls.append(i.get_attribute('href')) #Appending URLS

#Scraping the required data
counter=0 #Initializing the variable  
for i in urls:
    dr.get(i)
    counter+=1
    b=0
    c=40 # Variables b and c will be used as indices for slicing the data
    if counter==3:  # This if condition will change the default value of b and c in order to scrape only the first 20 data
        b=0
        c=20
    for a in dr.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')[b:c]: #Scraping Brand
        Brand.append(a.text)

    for a in dr.find_elements(By.XPATH,'//a[contains(@class,"IRpwTa")]')[b:c]: #Scraping Product
        Product.append(a.text)


    for a in dr.find_elements(By.XPATH,'//div[@class="_30jeq3"]')[b:c]: #Scraping Price 
        Price.append(a.text)
   
    for a in dr.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')[b:c]: #Scraping Discount
        Discount.append(a.text)
        

#Creating the DataFrame
Df_4=pd.DataFrame({'Brand':Brand,'Product':Product,'Price':Price,"Discount":Discount})
Df_4
```

```python
# Connecting to the driver
dr=webdriver.Chrome(executable_path=driver_path, options=options)
#Opening the website on automated brave browser
dr.get('https://www.azquotes.com/')


#Clicking the TOP QUOTES menu
dr.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a').click()


#Creating empty lists for the required data
Quote=[] 
Author=[] 
Type=[] 

#Adding delay to allow the website to load and update the content
time.sleep(4)


#Scraping the required data
for i in range(1,10): #Using for loop in order to run it 9 times
    
    for i in dr.find_elements(By.XPATH,'//a[@class="title"]'): 
        Quote.append(i.text)#Appending Quotes

    for i in dr.find_elements(By.XPATH,'//div[@class="author"]'): 
        Author.append(i.text)#Appending Authors


    for i in dr.find_elements(By.XPATH,'//div[@class="tags"]'): 
        Type.append(i.text)#Appending Types of Quotes
    
    dr.find_element(By.XPATH,'//li[@class="next"]').click() #Clicking on next button
    
    time.sleep(2) #Adding delay to allow the website to load and update the content

#Scraping the last page separately
for i in dr.find_elements(By.XPATH,'//a[@class="title"]'): 
    Quote.append(i.text)#Appending Quotes

for i in dr.find_elements(By.XPATH,'//div[@class="author"]'): 
    Author.append(i.text)#Appending Authors


for i in dr.find_elements(By.XPATH,'//div[@class="tags"]'): 
    Type.append(i.text)#Appending types of Quotes

#Creating the DateFrame
Df_8=pd.DataFrame({'Quote':Quote,' Author': Author,' Type Of Quotes': Type})
Df_8
```

```python
#Creating empty lists for the required data
Car_name=[] 
Price=[]

#Scraping the required data
for i in dr.find_elements(By.XPATH,'//h3[@class="subheader"]')[0:50]: #Using slice method to scrape desired data
    Car_name.append(i.text)#Appending Car Names

for i in dr.find_elements(By.XPATH,'//div[@class="postBody description e-content"]/p'): 
    Price.append(i.text)#Appending Price

# Filtering desired data in Price
f_price=[]
for i in Price:
    if i.startswith("Price: $") or i.endswith("Million"): # This condition states that the text which start with "Price: $" or ends with "Million" 
        f_price.append(i) #Appending desired data if condition is met true

#Creating the Date Frame
Df_10=pd.DataFrame({'Car name':Car_name,'Price': f_price})
Df_10
```
